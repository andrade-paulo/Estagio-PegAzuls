# Python e Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Asa

**Gráfico Cl x Alfa**

```python
dados_asa = pd.read_excel('polares_asa_e_empenagem_horizontal.xlsx', sheet_name='Dados da Aeronave', usecols="B:F", skiprows=3, nrows=51)
```

```python
dados_asa.head()
```

```python
plt.figure(figsize=(15,10))

plt.title('Cl pelo Alfa', fontsize=18)
plt.xlabel('Alfa', fontsize=15)
plt.ylabel('Cl', fontsize=15)

plt.xlim(-5, 20)
plt.ylim(0, 2.5)
plt.xticks(np.arange(-5, 21, 1))

plt.scatter(dados_asa['Alfa'], dados_asa['Cl'])
```
